```python
import copy

class Fruit:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
        print(f"Fruit constructor called for: {self.name}")

    def __del__(self):
        print(f"Fruit destructor called for: {self.name}")

    def __copy__(self):
        print(f"Fruit copy constructor called for: {self.name}")
        return type(self)(self.name, self.price, self.quantity)

    def __deepcopy__(self, memo):
        print(f"Fruit deep copy constructor called for: {self.name}")
        return type(self)(
            copy.deepcopy(self.name, memo),
            copy.deepcopy(self.price, memo),
            copy.deepcopy(self.quantity, memo)
        )

    def copy_assign(self, other):
        if self is not other:
            self.name = other.name
            self.price = other.price
            self.quantity = other.quantity
        print(f"Fruit copy assignment operator called for: {self.name}")

    def calculate_sum(self):
        return self.price * self.quantity

    def display(self):
        print(f"Fruit: {self.name}, Price: {self.price}, Quantity: {self.quantity}, Total: {self.calculate_sum()}")


class Vegetable:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
        print(f"Vegetable constructor called for: {self.name}")

    def __del__(self):
        print(f"Vegetable destructor called for: {self.name}")

    def __copy__(self):
        print(f"Vegetable copy constructor called for: {self.name}")
        return type(self)(self.name, self.price, self.quantity)

    def __deepcopy__(self, memo):
        print(f"Vegetable deep copy constructor called for: {self.name}")
        return type(self)(
            copy.deepcopy(self.name, memo),
            copy.deepcopy(self.price, memo),
            copy.deepcopy(self.quantity, memo)
        )

    def copy_assign(self, other):
        if self is not other:
            self.name = other.name
            self.price = other.price
            self.quantity = other.quantity
        print(f"Vegetable copy assignment operator called for: {self.name}")

    def calculate_sum(self):
        return self.price * self.quantity

    def display(self):
        print(f"Vegetable: {self.name}, Price: {self.price}, Quantity: {self.quantity}, Total: {self.calculate_sum()}")


if __name__ == "__main__":
    apple = Fruit("Apple", 2.5, 10)
    apple.display()

    carrot = Vegetable("Carrot", 1.8, 15)
    carrot.display()

    apple_copy = copy.copy(apple)
    apple_copy.display()

    carrot_copy = Vegetable("Temp", 0, 0)
    carrot_copy.copy_assign(carrot)
    carrot_copy.display()
```
